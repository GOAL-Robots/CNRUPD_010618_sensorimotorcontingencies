```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def oscillator(x,scale,p) :
    x = np.array(x)
    p = np.array(p)
    x = np.outer(x, np.ones(p.shape))
    return np.cos(p*5*np.pi*(x/scale-p))

scale = 1000.0
x = np.arange(scale)
p = np.random.rand(6)

y = oscillator(x,scale,p)

plt.plot(y)

y = oscillator(1,scale,p)
print y
```
